```python
import requests
import yaml
import json
import os
```

```python
# Load API key and endpoint from YAML config
with open("../../config/settings.yaml", "r") as file:
    config = yaml.safe_load(file)

API_KEY = config["news_api"]["api_key"]
ENDPOINT = config["news_api"]["endpoint"]

API_KEY  # just to verify it is loaded
```

```python
params = {
    "q": "India technology election",
    "language": "en",
    "pageSize": 20,
    "apiKey": API_KEY
}
```

```python
response = requests.get(ENDPOINT, params=params)

response.status_code
```

```python
data = response.json()

# Quick look at keys
data.keys()
```

```python
# Ensure raw data folder exists
os.makedirs("../../data/raw", exist_ok=True)

# Save raw data
with open("../../data/raw/news.json", "w", encoding="utf-8") as f:
    json.dump(data, f, indent=2)

print("✅ Raw news data saved to data/raw/news.json")
```
